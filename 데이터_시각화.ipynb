## 기초

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
from matplotlib import rc
```

```python
st1 = [100, 110, 90, 105, 95, 105, 90]
st2 = [70, 80, 80, 120, 90, 110, 105]

x = list(range(1, len(st1)+1))
```

```python
# 꺾은선 그래프
plt.title('two graph')
plt.plot(x, st1)
plt.plot(x, st2)
plt.legend(loc=2)
```

```python
# 막대 그래프
plt.title('graph 1')
plt.bar(x, st1)
plt.legend(loc=2)
```

```python
# 산점도 그래프
plt.title('graph 2')
plt.scatter(x, st2)
plt.legend(loc=2)
```

```python
import numpy as np
x_index = np.arange(len(st1))

plt.title('colorful graph')
plt.bar(x_index-0.2, st1, width=0.4, color='green',label='st1')
plt.bar(x_index+0.2, st2, width=0.4, color='blue',label='st2')
plt.grid(True)
plt.legend(loc=2)
plt.show()
```

```python
# 파이 그래프
ratio = [15, 21, 29, 35]
label = ['1', '2', '3', '4']

plt.title('pie chart')
plt.pie(x=ratio, labels=label, autopct='%.1f%%')
plt.show()
```

```python
explode = [0.1, 0, 0, 0]

plt.title('pie chart')
plt.pie(x=ratio, labels=label, autopct='%.1f%%', explode=explode)
plt.show()
```

## 실습

```python
import seaborn as sns
import pandas as pd

csData = pd.read_csv('/content/customerdata.csv')
csData.head()
```

```python
sns.countplot(data=csData, x='EMI', hue='CUSTTYPE')
```

```python
# 여러 플롯, 바+히스토그램
figure = plt.figure(figsize=(18,6))

ax1 = plt.subplot(1, 2, 1)
sns.distplot(csData.AVGPRICE, kde=False, ax=ax1)
ax2 = plt.subplot(1, 2, 2)
sns.distplot(csData.AVGPRICE, hist=False, ax=ax2)

figure.show()
```

```python
# 바 차트
sns.barplot(data = csData, x="EMI", y="AVGPRICE", hue = "CUSTTYPE", ci = "sd")
```

```python
# 박스 차트
sns.boxplot(data = csData, x="EMI", y="AVGPRICE")
```
