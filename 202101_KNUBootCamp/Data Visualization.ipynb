### 실습 I

P93 몸무게 변화 그래프 그리기 
weight.csv 업로드

weight.csv 파일을 읽어와서 날짜별로 몸무게를 계한사여 아래와 같은 몸무게 변화그래프를 그려보세요. (몸무게 계산은 human class 이용)

```
Human 클래스 정의하기
이름 Name, 나이 Age, 체중 Weight
init() : 이름 Name, 나이 Age, 체중 Weight 입력
eat() 섭취량, gym() 운동량 계산
write() 몸무게 출력 추가


다이어트 정보 파일 읽어오기 : weight.csv
읽어 온 데이터로 운동량리스트 생성하기 eat()/gim() 실행
날짜에 따른 몸무게 변화를 기록하는 딕셔너리 업데이트 

그래프로 시각화하기 - 선그래프 표현
```

````python
import pandas as pd

class human:
    def __init__(self, name, age, weight):
        self.Name = name
        self.Age = age
        self.Weight = weight

    def eat(self, meal):
        self.Weight += meal*0.01       

    def gym(self,kaloli):
        self.Weight -= kaloli* 0.002

    def write(self):        
        print(f"{self.Name} is {self.Age} years old and weights {self.Weight}Kg")

#메인

minsu = human("Min-su",16,55.0)
weight = {}     #key: 날짜, value : 몸무게가 되도록 설정

#1. csv파일 읽어오기

f = open("weight.csv")
data = pd.read_csv(f)


#2. eat/gym 조건에 따라 몸무게 계산
#3. 날짜별로 계산된 몸무게 저장

for row in data:    
    if row[2] =="eat":
        minsu.eat(int(row[3]))
        weight[row[0]] = minsu.Weight        
    if row[2] =="gym":
        minsu.gym(int(row[3]))
        weight[row[0]] = minsu.Weight      


print(weight)
#4. 그래프 출력
    
plt.title("Weight Progress state")
plt.xticks(rotation=45)
plt.ylabel("Weight",fontsize=12)
#plt.scatter(weight.keys(),weight.values())
plt.plot(list(weight.keys()),list(weight.values()), marker='^')  #scatter plot을 안그리고 마커로 해도 된다. 
plt.show()  
````

SciPy(과학기술계산을 위한 Python 라이브러리)를 이용하여 확률밀도함수 그리기

````python
from scipy.stats import norm
#평균이 10, 분산 3인 1000 개의 데이터 생성
data=norm.rvs(10.0,3,size=1000)
# density=True 이면 히스토그램을 그리면서 동시에 확률밀도값 반환
plt.hist(data, bins=20,density=True,alpha=0.5, color='b')

mu,std=norm.fit(data)
xmin,xmax=plt.xlim()#x의 최소, 최대값 구함
x=np.linspace(xmin,xmax,100)
p=norm.pdf(x,mu,std) # 평균과 표준편차로 부터 확률밀도 함수 생성
plt.plot(x,p,'r',linewidth=2)
````


2-1 대표적인 주기함수인 사인 함수를 사용하여 다음 그림과 같은 히트맵을 그려보도록 하자. 격자의 크기는 10x10 이며, 사인 함수의 입력값의 구간은 0에서 pi 사이이다.  


````python
import math
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, np.pi, 10)
y = np.linspace(0, np.pi, 10)

n = np.zeros(10*10).reshape(10, 10)
for i in range(10):
  for j in range(10):
    n[i, j] = np.sin(y[i])*np.sin(x[j])

plt.imshow(n)
plt.colorbar()
````

3 난수를 1,0000개 발생시켜 임의의 2차원 좌표를 생성해 그려보자. 이때 좌표의 x와 y 값은 표준정규분포를 따르도록 할 것이다. 그러면 생성된 좌표는 원점 (0,0)에 밀집한 모양을 가질 것이다. 


3-1 scatter() 함수 내에 불투명도를 의미하는 alpha 키워드 매개변수를 0.01로 두고 그림을 그려보자

````python
import matplotlib.pyplot as plt
import numpy as np

xData = np.random.randn(10000)  # 정규분포를 가지는 난수
yData = np.random.randn(10000)
plt.figure(figsize=(5,5))
plt.scatter(xData, yData, alpha=0.01)
plt.title('Random scatter plot')
plt.xlabel('x')
plt.ylabel('y')
````

3-2 alpha 키워드 매개변수를 0.1로 두고 색상을 ‘magenta’로 두어 심홍색 점이 진하게 보이도록 그려보자

````python
import matplotlib.pyplot as plt
import numpy as np

xData = np.random.randn(10000)  # 정규분포를 가지는 난수
yData = np.random.randn(10000)

plt.figure(figsize=(5,5))
plt.scatter(xData, yData, alpha=0.1, color='magenta')
plt.title('Random scatter plot')
plt.xlabel('x')
plt.ylabel('y')
````

##1 Tips 데이터로 seaborn 시작하기

식당에서 식사하고 지불한 팁

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns

sns.set_theme(style="darkgrid")
#tips  데이터셋을 불러옵시다. 
tips = sns.load_dataset("tips")
tips.head()
````

````python
# tips 데이터는 판다스 데이터프레임이다
tips.head(7)
````

````python
#relplot은 산점도와 선그래프 모두 그릴수 있는 함수 (kind='scatter'|'line')
sns.relplot(x = tips['total_bill'], y = tips['tip'], 
            hue = tips['sex'], col = tips['day'], row = tips['time'], sizes = (10, 100), kind = 'scatter')
````

````python
sns.relplot(x = tips['total_bill'], y = tips['tip'], 
            hue = tips['sex'], col = tips['day'], row = tips['time'], sizes = (10, 100), kind = 'line')
````

````python
tips.day.value_counts()
````

````python
tips.day.unique()
````

````python
tips.groupby('day').mean()
````

````python
grid = sns.FacetGrid(data = tips, col = 'day')
grid.map(sns.stripplot, 'time', 'tip')
````

````python
sns.relplot(x = 'total_bill', y = 'tip', hue = 'smoker', data = tips)
#hue 색상요소를 정하는 그룹화 요소
````

````python
sns.relplot(x = 'size', y = 'total_bill', data = tips)
````

````python
tips.shape
````

````python
#히스토그램
sns.distplot(tips['tip'], kde = False, bins = 10)
````

````python
#kde=True 키워드 인자를 통해서 가우시안 커널 밀도 추정gaussian kernel density estimation을 실선으로 그림
tips['tip_pct'] = 100 * tips['tip'] / tips['total_bill']
sns.distplot(tips['tip_pct'], kde = True, bins = 20)
````

선형회귀직선

````python
# regplot  (전체 비용에 대한 팁 비용의 상관관계 )
ax = sns.regplot(data = tips, x = tips['total_bill'], y = tips['tip'], ci = 98)
ax.set_xlabel('total_bill')
ax.set_ylabel('tip')
ax.set_title('Total Bill and Tip')
````

변수 사이의 관계를 알아보기에 편리한 쌍그래프

````python
sns.pairplot(tips)
````

시본 라이브러리의 pairplot() 함수에 kind='hist' 키워드 인자를 추가하여 왼쪽 그림과 같이 표시하여라. 다음으로 kind='kde'를 사용하여 쌍플롯을 그려 보아라.

````python
sns.pairplot(tips, kind = 'hist')
````

````python
sns.pairplot(tips, kind='kde')
````

##2  Anscombe’s quartet 데이터 셋

1973년 통계학자인 프란시스 앤스컴Francis Anscombe이 데이터 분석에 앞서 시각화의 중요성과 특이치 값의 영향을 보여주기 위해 만든 데이터셋

````python
import seaborn as sns

anscombe = sns.load_dataset("anscombe")
print(anscombe.head())
````

````python
# 데이터셋 구성 확인해보기

anscombe.dataset.value_counts()
````

````python
sns.relplot(data = anscombe, x = 'x', y = 'y', col = 'dataset')
````

````python
anscombe.query("dataset == 'I'")
````

lmplot은 regplot과 유사한 선형 회귀 직선을 구하는 기능과 함께 FacetGrid라고 하는 화면에 보기 좋은 격자간격을 생성하는 인터페이스를 결합한 함수


````python
sns.lmplot(data = anscombe, x = 'x', y = 'y', col = "dataset")
````

비선형 함수를 사용

````python
#order=2 비선형함수
sns.lmplot(data = anscombe.query("dataset == 'II'"), x = 'x', y = 'y', order = 2)
````

이상치 존재

lmplot() 함수를 robust=True 키워드 인자를 사용하면 이상치에 덜 민감한 강건한 모델을 만듦

````python
sns.lmplot(data = anscombe.query("dataset == 'III'"), x = 'x', y = 'y', robust = True, ci = None)
````

````python
sns.lmplot(data = anscombe.query("dataset == 'IV'"), x = 'x', y = 'y', robust = True, ci = None)
````
