*   두 수 중 작은 수를 반환하는 함수 get_min(num1,num2)
*   거듭 제곱을 계산하는 함수 power(num1,num2)항목 추가

````python
def square(num):
  return num*num

def get_max(num1, num2):
  if num1 > num2:
    return num1
  elif num1 < num2:
    return num2
  else:
    print('두 값이 같습니다')
    return num1

def get_min(num1, num2):
  if num1 < num2:
    return num1
  elif num1 > num2:
    return num2
  else:
    print('두 값이 같습니다')
    return num1

def power(num1, num2):
  result = 1
  for i in range(0, num2):
    result *= num1
  return result

num = 7
num1 = 5
num2 = 3

print(square(num))
print(get_max(num1,num2))
print(get_min(num1,num2))
print(power(num1,num2))
````

####연산자오버로딩

````python
class Method:
    def  __init__(self, obj):
        self.obj = obj
    def  _value_check(self, value):
        return value if isinstance(value, int) else int(value)

    def __add__(self, value):
        value = self._value_check(value)
        self.obj = [x+value for x in self.obj]
        return self

    def __sub__(self, value):
        value = self._value_check(value)
        self.obj = [x-value for x in self.obj]
        return self

    def __mul__(self, value):
        value = self._value_check(value)
        self.obj = [x*value for x in self.obj]
        return self

    def __truediv__(self, value):
        value = self._value_check(value)
        self.obj = [int(x/value) for x in self.obj]
        return self

m = Method([1,2,3,4,5])
m = m+3
print(m.obj)
````

````python
import csv
import matplotlib.pyplot as plt

class human:
    def __init__(self, name, weight):
        self.name = name
        self.weight = weight
        self.weight_list = {}

    def eat(self, today, meal):
        self.weight += meal * 0.01
        self.weight_list[today] = self.weight

    def gym(self, today, kaloli):
        self.weight += kaloli * 0.002
        self.weight_list[today] = self.weight

    def write(self):
        for day, weight in self.weight_list.items():
            print(f'date = : {day}, weight = {weight:.2f}')

class woman(human):
    def __init__(self, name, weight, height):
      super().__init__(name, weight)
      self.height = height

    def bmi_calc(self):
      bmi = self.weight / (self.height * self.height)
      if bmi < 18.5:
        print('저체중입니다')
      elif bmi < 25:
        print('정상체중입니다')
      elif bmi < 30:
        print('과체중입니다')
      elif bmi < 40:
        print('비만입니다')
      else:
        print('고도비만입니다')
        
    def showInfo(self):
      print(f'키 : {self.height}cm 체중 : {self.weight}kg')
      bmi_calc()

minsu = woman('Min-su', 50, 160)
print(minsu.showInfo())

````

### [실습] txt파일 다루기 (P394)

1. "매수종목.txt" 파일을 만들고 아래와 같이 입력한다
```
005930 삼성전자
005380 현대차
035420 NAVER

2. 파일의 내용을 매수종목번호를 키, 종목이름을 값으로 읽어들여 화면에 출력한다. 
```
{'005930': '삼성전자', '005380': '현대차', '035420': 'NAVER'}
```

````python
#파일 쓰기
f = open('매수종목.txt', mode = 'w', encoding = 'utf-8')
f.write('005930 삼성전자\n')
f.write('005380 현대차\n')
f.write('035420 NAVER\n')
f.close()

#파일 읽기
f = open('매수종목.txt', encoding = 'utf-8')
lines = f.readlines()
data = {}

for line in lines:
  line = line.strip()
  k, v = line.split()
  data[k] = v

print(data)
f.close()
````

### [미니 프로젝트] 다이어트 프로그램 (PP.396~398)

1. 고객데이터를 랜덤하게 만들어 파일 'info.txt'에 저장한다.
2. 고객데이터를 읽고 bmi를 계산해서 출력한다.

 BMI = (몸무게 / 키의 제곱)*10000

> *   25이상 비만
*   23이상 25미만 과체중
*   18.5이상 23미만 정상체중
*   18.5미만 저체중



````python
#info.txt 만들기
import random
alphabet = list('가나다라마바사아자차카타파하')

with open('info.txt', 'w') as file:
  for i in range(1000):
    name = random.choice(alphabet) + random.choice(alphabet)
    weight = random.randrange(40, 100)
    height = random.randrange(140, 200)
    file.write('{}, {}, {}\n'.format(name, weight, height))

````

````python
#고객데이터파일을 읽고 bmi계산 출력
with open('info.txt', 'r') as file:
  for line in file:
    (name, weight, height) = line.strip().split(', ')

    if (not name) or (not weight) or (not height): continue
    else:
      bmi = int(int(weight) / (int(height) * int(height)) * 10000)
      result = ''
      if bmi < 18.5:
        result = '저체중'
      elif bmi < 23:
        result = '정상체중'
      elif bmi < 25:
        result = '과체중'
      else:
        result = '비만'

      print('이름 : {}'.format(name))
      print('체중 : {}'.format(weight))
      print('신장 : {}'.format(height))
      print('BMI : {}'.format(bmi))
      print('결과 : {}\n'.format(result))
````
