##merge : DataFrame.merge(right,how='inner',on=None)
```

how : 결합의 방식 'left','right','inner','outer' 가능 (inner가 default)
```

# 실습 IV

stock_price.csv / stock_valuation.csv 파일을 이용하여 데이터프레임을 변형시켜 볼 것이다. 
```
EPS = 당기순이익 / 발행주식수 : 1주당 벌어들인 금액 : 주단순이익
BPS = ( 자산통계 - 부채총계 ) / 발행 주식수 : 주당순자산
PER = 현재주가 / EPS  : 현재 주가의 평가 정도 ; 주가순익비율
PBR = 현재주가 / BPS : 주가순자산비율
```

````python
import pandas as pd

df1 = pd.read_csv("stock_price.csv")
df2 = pd.read_csv("stock_valuation.csv")

print(df1, '\n')
print(df2)
````

````python
pd.merge(df1, df2)
````

````python
# 둘다 공통으로 들어있는 id만 추출되므로 outer조인을 해보아야 겠다. 
pd.merge(df1, df2, how = 'outer', on = 'id')
````

````python
# 기업 이름의 항목이 겹치므로 좌우 두개의 데이터프레임에서 대표키가 될 항목을 정한 우 우선순위를 결정해본다. 
pd.merge(df1, df2, how = 'left', left_on = 'stock_name', right_on = 'name')
````

````python
pd.merge(df1, df2, how = 'right', left_on = 'stock_name', right_on = 'name')
````

````python
pd.merge(df1, df2, how = 'outer', left_on = 'stock_name', right_on = 'name')
````
